## filter, arrange, select, mutate, summarise

```r
install.packages("nycflights13")
library(nycflights13)
flights[1:5,]
```

## filter

```r
filter(flights, month ==1)
```

```r
filter (flights, month ==11 | month == 12)
```

```r
# This will select every row where x is one of the values in y 
filter(flights, month %in% c(10,12))
```

## arrange
> arrange() works similarly to filter() except that instead of selecting rows it changes thier order.
it takes a data frame and set of column names **to order by**.


```r
arrange(flights, year, month, day)
```

```r
## default = asc
arrange(flights, desc(dep_delay))
```

## Select 
rapidly zoom in on a useful subset using operations based on the names of the 
>> starts_with("abc")
>> ends_with("xyz")
>> contains("ijk")


```r
select(flights, year, month, day)[1:5,]
```

```r
select(flights, year:day)[1:5,]
```

```r
select(flights, -(year:day))
```

## mutate()
add variable

```r
flights_sml <- select(flights, year:day, ends_with("delay"), distance, air_time)
mutate(flights_sml, 
      gain = dep_delay -arr_delay, 
      speed = distance/air_time *60)
```

## summarise

```r
summarise(flights, delay = mean(dep_delay, na.rm=TRUE))
```

```r
by_day <-group_by(flights, year, month, day)
summarise(by_day, delay = mean(dep_delay, na.rm=TRUE))
```

## Improve readability using pipe operators
x%>%f(y) turns into f(x,y)

```r
by_dest <-group_by(flights, dest)
delay <- summarise(by_dest, count=n(), dist = mean(distance, na.rm = TRUE), delay = mean(arr_delay, na.rm=TRUE))
delay <-filter(delay, count >20, dest !="NHL")
delay
```

```r
delays <- flights %>%
group_by(dest) %>% 
summarise(count = n(), dist= mean(distance, na.rm = TRUE), delay = mean(arr_delay, na.rm = TRUE))%>% 
filter(count >20, dest !="NHL")
delays
```

```r
flights %>%
group_by(origin) %>%
        summarize(count = n(), dist = mean(distance, na.rm=TRUE))
```

```r
temp2 <- flights %>%
group_by(origin) %>%
        summarize(count = n(), dist = mean(distance, na.rm=TRUE))
ggplot(data = temp2, aes(x = origin, y = count)) + geom_bar(stat='identity')
```
